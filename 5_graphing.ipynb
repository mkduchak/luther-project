```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
playerdata = pd.read_pickle('data/playerdatafull')
advstats = pd.read_pickle('data/advstats')
```

```python
playervorp = playerdata[['player','avgVORP']]
top10players = playervorp.sort_values('avgVORP',ascending=False).head(10).reset_index(drop=True)
```

```python
fig, ax = plt.subplots(1, figsize=(10, 5))
y_pos = np.arange(10)
ax.set_yticks(y_pos)
ax.set_yticklabels(top10players['player'], fontsize = 12)
ax.set_ylabel('Players',fontsize = 16,rotation = 0,labelpad=30)
ax.set_xlabel('Career Avg VORP',fontsize=16)
ax.barh(y_pos, top10players['avgVORP'], color='g', align='center')
ax.invert_yaxis();
```

```python
bottom10players = playervorp.sort_values('avgVORP',ascending=True).head(10).reset_index(drop=True)
```

```python
fig, ax = plt.subplots(1, figsize=(10, 5))
y_pos = np.arange(10)
ax.set_yticks(y_pos)
ax.set_yticklabels(bottom10players['player'], fontsize = 12)
ax.set_ylabel('Players',fontsize = 16,rotation = 0,labelpad=30)
ax.set_xlabel('Career Avg VORP',fontsize=16)
ax.barh(y_pos, bottom10players['avgVORP'], color='r', align='center')
ax.invert_yaxis();
```
